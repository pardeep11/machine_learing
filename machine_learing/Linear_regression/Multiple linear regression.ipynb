# Multiple Linear Regression in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/home/pardeep/Boston.csv')
```

```python
dataset
```

```python
dataset.head()
```

```python
#Set up the dependent and the independent variables
x=pd.DataFrame(dataset.iloc[:,:-1])
y=pd.DataFrame(dataset.iloc[:,-1])
```

```python
x.head()  #independent variable
```

```python
y.head()#dependent
```

```python
#Divide the data into train and test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
#Have a glance at the shape of the train and test sets:
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

```

```python
#Train the algorithm:
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Having a look at the coefficients that the model has chosen:
v=pd.DataFrame(regressor.coef_,index=['Co-efficient']).transpose()
w=pd.DataFrame(x.columns,columns=['Attribute'])
```

```python
#Concatenating the DataFrames to compare
coeff_df=pd.concat([w,v],axis=1,join='inner')
coeff_df
```

```python
#Comparing the predicted value to the actual value:
y_pred=regressor.predict(x_test)
y_pred=pd.DataFrame(y_pred,columns=['predicted'])
y_pred.head()

```

```python
#Evaluate the algorithm
#Evaluate the algorithm
from sklearn import metrics
print('Mean Squared Error',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
